```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['figure.figsize'] = [10, 6]
```

```python
x=np.linspace(-3,3,50)
V=1000*x*x
Va=V*(1-0.04*x)
n=np.arange(0,5,1)
E_arm=2000*(n+1/2)
plt.plot(x,V)
plt.plot(x,Va)
plt.barh(E_arm, 3, height=100)
χe=0.02
E_anharm=E_arm*(1-χe*(n+1/2))
plt.barh(E_anharm, -3, height=100)
```

```python
B=40
lev=5
J=np.arange(0,lev,1)
Ej=B*J*(J+1)
#plt.plot(x,Va)
for i in range(2):
    plt.barh(E_anharm[i]+Ej,3,height=20)
for i in range(lev-1):
    print("{},{} : {}".format( i, i+1, Ej[i+1]-Ej[i]))
    
for i in range(lev-1):
    da=E_anharm[0]+Ej[i]
    a=E_anharm[1]+Ej[i+1]
    plt.annotate("",(2.0+i*0.2, a),(2.0+i*0.2, da), arrowprops={'arrowstyle':'->'})
    da=E_anharm[0]+Ej[i+1]
    a=E_anharm[1]+Ej[i]
    plt.annotate("",(1.0-i*0.2, a),(1.0-i*0.2, da), arrowprops={'arrowstyle':'->'})
```

```python
ν0=2000
B=12
J=np.arange(0,15,1)
Ej=B*J*(J+1)
rR=ν0+Ej[1:]-Ej[:-1]
rP=ν0+Ej[:-1]-Ej[1:]
Ir=J*np.exp(-J*(J+1)/20)
plt.bar(rR,Ir[1:],width=3)
plt.bar(rP,Ir[1:],width=3)
```

$\omega=\sqrt {\frac k \mu}$

$B=\frac{\hbar^2}{2\mu}$

```python
M12=12*16/(12+16)
M13=13*16/(13+16)
ν0_13=ν0*np.sqrt(M12/M13)
B_13=B*(M12/M13)

Ej_13=B_13*J*(J+1)
rR_13=ν0_13+Ej_13[1:]-Ej_13[:-1]
rP_13=ν0_13+Ej_13[:-1]-Ej_13[1:]
Ir_13=J*np.exp(-J*(J+1)/20)*0.06
plt.bar(rR,Ir[1:],width=3, color='black')
plt.bar(rP,Ir[1:],width=3, color='black')
plt.bar(rR_13,Ir_13[1:],width=3, color='red')
plt.bar(rP_13,Ir_13[1:],width=3, color='red')
```

```python
M35=35*1/(35+1)
M37=37*1/(37+1)
ν0_37=ν0*np.sqrt(M35/M37)
B_37=B*(M35/M37)

Ej_37=B_37*J*(J+1)
rR_37=ν0_37+Ej_37[1:]-Ej_37[:-1]
rP_37=ν0_37+Ej_37[:-1]-Ej_37[1:]
Ir_37=J*np.exp(-J*(J+1)/20)*0.35
plt.bar(rR,Ir[1:],width=3, color='black')
plt.bar(rP,Ir[1:],width=3, color='black')
plt.bar(rR_37,Ir_37[1:],width=3, color='red')
plt.bar(rP_37,Ir_37[1:],width=3, color='red')
plt.xlim(1800,2200)
```
